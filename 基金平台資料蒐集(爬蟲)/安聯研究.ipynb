```python
def find_content(links):
    
    s_list = []

    res2 = requests.get(links)
    soup2 = BeautifulSoup(res2.text, "html.parser")
    temp_content = soup2.select(".l-grid-row p")
    for data in temp_content:
        if data.text.__contains__("限時") or data.text.__contains__("資料來源"):
            continue            
        s_list.append(data.text.strip())

    return "".join(s_list)
```

```python
df1["內容"]=df1["內容連結"].apply(find_content)
```

```python
import nltk
import math
import string
from nltk.corpus import stopwords
from collections import Counter
from nltk.stem.porter import *
from sklearn.feature_extraction.text import TfidfVectorizer
import jieba
```

```python
def jieba_tool(content):
    list4=[]

    f = open(r"stopwords_TW.txt",'r',encoding = 'UTF-8')
    stopwords_n = f.readlines()
    f.close()  
    stopwords = [sw.strip().replace('\n','') for sw in stopwords_n]
    jieba.set_dictionary('dict_TW.txt')
    wordlist=[]


    words=jieba.cut(content, cut_all=False)
    for word in words:
        wordlist.append(word.replace(" ",""))
        no_punctuation=[x for x in wordlist if not x in stopwords]
    return Counter(no_punctuation)
```

```python
df1["斷詞"]=df1["內容"].apply(jieba_tool)
```

```python
import datetime

def get_week_of_month(year, month, day):
    begin = int(datetime.date(year, month, 1).strftime("%W"))
    end = int(datetime.date(year, month, day).strftime("%W"))

    return end - begin + 1
```

```python
def tf(word, count):   #tf-idf運算套件
    return count[word] / sum(count.values())
def n_containing(word, count_list):
    return sum(1 for count in count_list if word in count)
def idf(word, count_list):
    return math.log(len(count_list) / (1 + n_containing(word, count_list)))
def tfidf(word, count, count_list):
    return tf(word, count) * idf(word, count_list)


list1=[]
list2=[]

def tfidf_count(countlist):
    for i, count in enumerate(countlist):
            scores = {word: tfidf(word, count, countlist) for word in count}
            sorted_words = sorted(scores.items(), key = lambda x: x[1], reverse=True)   
            for word, score in sorted_words:
                list1.append("Word: {}, TF-IDF: {}".format(word,score))
            list1_copy=list(list1)
            list2.append(list1_copy)
    return(list2)
```
