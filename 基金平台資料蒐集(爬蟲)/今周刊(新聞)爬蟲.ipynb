```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

titles=[]
links=[]
dates=[]

url = 'https://www.businesstoday.com.tw/catalog/183010/photo/page/7'
                                                                                                                                                                                                                                                                                                                                                                                                               
res=requests.get(url)

soup = BeautifulSoup(res.text, 'html.parser') 

time = soup.select(".article__block-date")
title = soup.select(".article__block-title")
link = soup.find_all('a', attrs={"class": "article__block-box"})


for each_date in time:
    dates.append(each_date.text.strip())
for each_title in title:
    titles.append(each_title.text)
for each_link in link:
    links.append(each_link['href'])
        
df = pd.DataFrame(
{
    '日期':dates,
    '標題':titles,
    '內容連結':links
})
```

```python
df
```

```python
def find_content(link):
    s_list = []

    res2 = requests.get(link)
    soup2 = BeautifulSoup(res2.text, "html.parser")
    link_id=link.split("/")[7]
    temp_content = soup2.select("#article__page__"+link_id)
    for data in temp_content:
        s_list.append(data.text.replace("\n\xa0\n","").strip())

    return "".join(s_list)
```

```python
df["內容"]=df["內容連結"].apply(find_content)
```

```python
import nltk
import math
import string
from nltk.corpus import stopwords
from collections import Counter
from nltk.stem.porter import *
from sklearn.feature_extraction.text import TfidfVectorizer
import jieba
```

```python
def jieba_tool(content):
    list4=[]

    f = open(r"stopwords_TW.txt",'r',encoding = 'UTF-8')
    stopwords_n = f.readlines()
    f.close()  
    stopwords = [sw.strip().replace('\n','') for sw in stopwords_n]
    jieba.set_dictionary('dict_TW.txt')
    wordlist=[]


    words=jieba.cut(content, cut_all=False)
    for word in words:
        wordlist.append(word.replace(" ",""))
        no_punctuation=[x for x in wordlist if not x in stopwords]
    return no_punctuation
```

```python
df["斷詞結果"]=df["內容"].apply(jieba_tool)
```

```python
import datetime

def get_week_of_month(year, month, day):
    begin = int(datetime.date(year, month, 1).strftime("%W"))
    end = int(datetime.date(year, month, day).strftime("%W"))

    return end - begin + 1
```

```python
week=[]
for time in df["日期"]:
    y=int(time[0:4])
    m=int(time[5:7])
    d=int(time[8:10])
    week_id=get_week_of_month(y, m, d)
    week.append(time[0:4]+"-"+time[5:7]+"-"+str(week_id))
df["週期ID"]=week
df
```

```python
df.to_csv("今周刊(新聞)斷詞.csv",encoding="utf_8_sig")
```
